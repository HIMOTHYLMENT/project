## Financial irregularities

### Subtask:
Identify accounts with potentially abnormal opening or closing balances for indigent applicants.


**Reasoning**:
Examine the 'Opening Balance' and 'Closing Balance' columns to identify potentially abnormal values by first looking at descriptive statistics.



```python
print("Descriptive statistics for 'Opening Balance':")
display(df['Opening Balance'].describe())

print("\nDescriptive statistics for 'Closing Balance':")
display(df['Closing Balance'].describe())
```

**Reasoning**:
Visualize the distribution of 'Opening Balance' and 'Closing Balance' using box plots to better identify outliers that might indicate abnormal balances.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a box plot for 'Opening Balance'
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
sns.boxplot(y=df['Opening Balance'])
plt.title('Box Plot of Opening Balance')
plt.ylabel('Opening Balance')

# Create a box plot for 'Closing Balance'
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
sns.boxplot(y=df['Closing Balance'])
plt.title('Box Plot of Closing Balance')
plt.ylabel('Closing Balance')

plt.tight_layout()
plt.show()
```

**Reasoning**:
Identify rows where 'Opening Balance' or 'Closing Balance' are significantly different from the median (or a reasonable threshold based on the box plots) to flag potentially abnormal balances. A simple approach is to consider values beyond a certain percentile or a fixed threshold based on the observed range. Given the box plots show many values at 0 and some large outliers, let's identify values far from the median (0 in this case) or beyond the observed IQR range multiplied by a factor.



```python
# Calculate IQR for both columns
Q1_opening = df['Opening Balance'].quantile(0.25)
Q3_opening = df['Opening Balance'].quantile(0.75)
IQR_opening = Q3_opening - Q1_opening

Q1_closing = df['Closing Balance'].quantile(0.25)
Q3_closing = df['Closing Balance'].quantile(0.75)
IQR_closing = Q3_closing - Q1_closing

# Define outlier bounds using 1.5 * IQR rule
lower_bound_opening = Q1_opening - 1.5 * IQR_opening
upper_bound_opening = Q3_opening + 1.5 * IQR_opening

lower_bound_closing = Q1_closing - 1.5 * IQR_closing
upper_bound_closing = Q3_closing + 1.5 * IQR_closing

# Identify rows with values outside these bounds
abnormal_balances_df = df[
    (df['Opening Balance'] < lower_bound_opening) |
    (df['Opening Balance'] > upper_bound_opening) |
    (df['Closing Balance'] < lower_bound_closing) |
    (df['Closing Balance'] > upper_bound_closing)
]

print("Rows with potentially abnormal Opening or Closing Balances:")
if not abnormal_balances_df.empty:
    display(abnormal_balances_df[['Account Number', 'Applicant Id Number', 'Opening Balance', 'Closing Balance']])
else:
    print("No potentially abnormal balances identified based on the 1.5*IQR rule.")
```
